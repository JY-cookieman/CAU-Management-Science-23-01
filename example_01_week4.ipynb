```python
import matplotlib.pyplot as plt
import numpy as np

# increase default figure and font sizes for easier viewing
plt.rcParams['figure.figsize'] = (8, 8)
plt.rcParams['font.size'] = 14
```

Constraints and moving objective function to the north

```python
plt.plot([0,200],[200,0],color="red")
plt.plot([0,174],[261,0],color="blue")
plt.plot([0,240],[180,0],color="green")

plt.plot([0,12],[14,0],"--",color="red") #z=4200
plt.plot([0,12*10],[14*10,0],"--",color="red") #z=42000
plt.plot([0,12*15],[14*15,0],"--",color="red") #z=63000
plt.plot([0,12*20],[14*20,0],"--",color="red") #z=84000
```

Find the coordinates of intersection of two constraints

```python
a = np.array([[1,1],[9,6]])
b = np.array([200,1566])
x = np.linalg.solve(a,b)
print(x)
```

```python
plt.plot([0,200],[200,0],color="red")
plt.plot([0,174],[261,0],color="blue")
plt.plot([0,240],[180,0],color="green")

plt.plot([0,188.85714],[220.3333,0],"--",color="black") #z=66100
plt.scatter(x[0],x[1],color="red") #z=66100

plt.xlim(0,260)
plt.ylim(0,260)
plt.show()
```
