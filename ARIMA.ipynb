```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
shampoo_train = shampoo[0:25]
shampoo_test = shampoo[25:36]
```

```python
shampoo_model = ARIMA(shampoo_train, order=(3,1,2), trend='t')
shampoo_model_fit = shampoo_model.fit()
shampoo_model_fit.aic
shampoo_forecast = shampoo_model_fit.forecast(steps=11)
shampoo_forecast
```

```python
np.sqrt(mean_squared_error(shampoo_test, shampoo_forecast))
```

```python
p_values = range(0,4)
d_values = range(0,4)
q_values = range(0,4)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for p in p_values:
    for d in d_values:
        for q in q_values:
            order = (p,d,q)
            train, test = shampoo[0:25], shampoo[25:36]
            predictions = list()
            for i in range(len(test)):
                try:
                    model = ARIMA(train, order=order, trend='t')
                    model_fit = model.fit()
                    pred_y = model_fit.forecast()
                    predictions.append(pred_y)
                    print(len(test), len(predictions), pred_y)
                    error = np.sqrt(mean_squared_error(test, predictions))
                    print('ARIMA%s RMSE = %.2f'% (order, error))
                except:
                    continue
```
